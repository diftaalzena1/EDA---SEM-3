### **Deskripsi Kasus: Analisis Laporan Kebahagiaan Dunia**

**Pendahuluan:**
Dataset Kebahagiaan Dunia (World Happiness Report) ini mengevaluasi tingkat kebahagiaan di 155 negara berdasarkan data dari Gallup World Poll. Survei ini menggunakan *Cantril ladder* untuk mengukur kebahagiaan, di mana responden diminta untuk menilai kehidupan mereka pada skala 0 (terburuk) hingga 10 (terbaik). Hasil laporan ini dimanfaatkan oleh pemerintah dan organisasi untuk memahami faktor-faktor yang mempengaruhi kesejahteraan masyarakat serta menjadi acuan dalam penyusunan kebijakan.

**Tujuan Analisis:**
Analisis ini bertujuan untuk mengidentifikasi faktor-faktor utama yang memengaruhi tingkat kebahagiaan di berbagai negara, memahami perbedaan antar negara, serta memantau perubahan tren kebahagiaan dari waktu ke waktu.

### **Deskripsi Variabel dalam Dataset:**

1. **Country** *(Nominal)*: Nama negara yang dinilai dalam survei.
2. **Region** *(Nominal)*: Kawasan geografis tempat negara tersebut berada.
3. **Happiness Rank** *(Ordinal)*: Peringkat kebahagiaan negara berdasarkan skor total.
4. **Happiness Score** *(Rasio)*: Skor rata-rata kebahagiaan, diukur dengan skala 0-10.
5. **Economy (GDP per Capita)** *(Rasio)*: Kontribusi pendapatan ekonomi per kapita terhadap kebahagiaan.
6. **Family** *(Rasio)*: Dukungan sosial atau keakraban keluarga yang berpengaruh pada kebahagiaan.
7. **Health (Life Expectancy)** *(Rasio)*: Harapan hidup sebagai indikator kesehatan dan kebahagiaan.
8. **Freedom** *(Rasio)*: Kebebasan individu dalam membuat pilihan hidup.
9. **Trust (Government Corruption)** *(Rasio)*: Tingkat kepercayaan masyarakat terhadap pemerintah dan rendahnya korupsi.
10. **Generosity** *(Rasio)*: Tingkat kedermawanan atau sikap saling memberi di masyarakat.

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Misalkan df_2015 adalah DataFrame Anda yang sudah di-load
df_sorted = df_2015.sort_values(by='Happiness Score', ascending=False)
highest_df = df_sorted.head(10)
lowest_df = df_sorted.tail(10)

# Membuat subplot dengan dua grafik
fig = make_subplots(rows=1, cols=2, subplot_titles=("Top 10 Countries by Happiness Score", "Bottom 10 Countries by Happiness Score"))

# Menambahkan grafik untuk 10 negara dengan skor tertinggi
fig.add_trace(
    go.Scatter(
        x=highest_df["Country"], 
        y=highest_df["Happiness Score"], 
        mode='markers+lines+text',
        text=highest_df["Country"],  
        textposition='top center',   
        textfont=dict(size=10),      
        name='Highest Scores'
    ),
    row=1, col=1
)

# Menambahkan grafik untuk 10 negara dengan skor terendah
fig.add_trace(
    go.Scatter(
        x=lowest_df["Country"], 
        y=lowest_df["Happiness Score"], 
        mode='markers+lines+text',
        text=lowest_df["Country"],  
        textposition='top center',   
        textfont=dict(size=10),      
        name='Lowest Scores'
    ),
    row=1, col=2
)

# Menyusun tata letak dan tampilkan grafik
fig.update_layout(
    height=800, 
    width=1300, 
    title_text="Highest & Lowest Countries Score",
    xaxis_title='Country',
    yaxis_title='Happiness Score',
    xaxis_tickangle=-45  
)

fig.show()
```

#### 11. Bar Chart

```python
# Bar chart for showing mean happiness score for each region

happiness_per_region = df_2015.groupby('Region')['Happiness Score'].mean().reset_index()
happiness_per_region.columns = ['Region', 'Happiness_Score']

sns.set_palette("Set1")

plt.figure(figsize=(12, 8))

ax = sns.barplot(data=happiness_per_region, x="Region", y="Happiness_Score")

plt.xticks(rotation=80)
plt.title("Mean Happiness Score for Each Region (2015)")
plt.ylabel("Happiness Score")

ax.set(xlabel=None)
for i in ax.containers:
    ax.bar_label(i, fmt='%.2f')

plt.show()
```

**Interpretasi:**

Bar chart ini menunjukkan rata-rata skor kebahagiaan untuk masing-masing wilayah dalam dataset `df_2015`. Setiap batang mewakili rata-rata skor kebahagiaan di suatu wilayah, dengan warna yang berbeda untuk setiap wilayah.

Dari chart ini, saya bisa melihat perbedaan dalam skor kebahagiaan rata-rata di berbagai wilayah. Misalnya, batang untuk wilayah Australia and New Zealand lebih tinggi, ini menunjukkan bahwa negara-negara di wilayah tersebut memiliki rata-rata skor kebahagiaan yang lebih tinggi dibandingkan dengan wilayah lain. Sebaliknya, wilayah yang memiliki batang lebih pendek, yakni seperti Sub-Saharan Africa menunjukkan rata-rata skor kebahagiaan yang lebih rendah.
